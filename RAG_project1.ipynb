```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Direct link to the PDF
url = 'https://www.indiabudget.gov.in/doc/budget_speech.pdf'

# Load the PDF
loader = PyPDFLoader(url)
data = loader.load()

# Split the text into manageable chunks for the LLM
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
docs = text_splitter.split_documents(data)

print(f"Loaded {len(data)} pages and created {len(docs)} chunks.")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma

# Initialize embeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Create the vector store
vectorstore = Chroma.from_documents(documents=docs, embedding=embeddings)

# Create a retriever object
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
import torch
from transformers import pipeline, BitsAndBytesConfig
from langchain_huggingface import HuggingFacePipeline

model_id = "tiiuae/falcon-7b"

# Configuration to compress the model so it fits in memory
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Initialize the pipeline
text_generation_pipeline = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"quantization_config": quant_config},
    max_new_tokens=400,
    device_map="auto"
)

llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

# Fixed typos: 'promp' -> 'prompt', 'StrOutputPara' -> 'StrOutputParser'
prompt_template = """
<|system|>
Answer the question based on the context provided below. If you don't know, say you don't know.

Context:
{context}

</s>
<|user|>
{question}
</s>
<|assistant|>
"""

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_template,
)

# Create the final RAG chain
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# Define llm_chain using existing prompt and llm objects
llm_chain = prompt | llm | StrOutputParser()

# Redefine rag_chain
rag_chain = {"context":retriever,"question":RunnablePassthrough()} | llm_chain
```

```python
question = "who introduced the budget"
```

```python
rag_chain.invoke(question)
```

```python
response = rag_chain.invoke(question)
response = response.replace('</s>', '')
print("Responce:",response)
```
